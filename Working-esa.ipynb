```python
#noticed that there is usually only one entry per individual each day
#creating new index (fitbase id + date) for each entry
#this will link the three dataframes together
activity_newid = activity['Id'].astype(str) + '_' + activity['ActivityDate'].astype(str)
activity_newid
```

```python
sleep_newid = sleep['Id'].astype(str) + '_' + sleep['SleepDay'].astype(str)
sleep_newid
```

```python
sleepmins_newid = sleepmins['Id'].astype(str) + '_' + sleepmins['date'].dt.date.astype(str)
sleepmins_newid
```

```python
#adding new_ids into dataframes
activity = pd.concat([activity, activity_newid], axis = 1)
```

```python
activity = activity.rename(columns= {0: "newid"})
```

```python
sleep = pd.concat([sleep, sleep_newid], axis = 1)
```

```python
sleep = sleep.rename(columns= {0: "newid"})
```

```python
sleepmins = pd.concat([sleepmins, sleepmins_newid], axis = 1)
```

```python
sleepmins = sleepmins.rename(columns= {0: "newid"})
```

```python
#finding out earliest and latest minute of sleep for each day 
#(i.e. the time they fell asleep and time they woke up)
start_end = sleepmins.groupby('newid')['date'].agg(['min', 'max'])
start_end
```

```python
start_end['min'] = start_end['min'].dt.strftime("%H:%M:%S")
```

```python
start_end['max'] = start_end['max'].dt.strftime("%H:%M:%S")
```

```python
start_end = start_end.rename(columns = {'min':'SleepStart', 'max': 'SleepEnd'})
```

```python
start_end
```

```python
#merging sleepmins and start_end together
sleepmins = pd.merge(sleepmins, start_end, on= 'newid')
sleepmins
```

```python
#dropping duplicates so each unique sleep entry with its start and end date for joining later
sleepmins = sleepmins.drop(columns=['Id','date','value','logId'])
sleepmins = sleepmins.drop_duplicates()
sleepmins
```

```python
#merging sleep and sleepmins dataframes together
sleep = pd.merge(sleep, sleepmins)
sleep
```

```python
#merging activity and sleep dataframes together
activity_sleep = pd.merge(activity, sleep)
```

```python
activity_sleep = activity_sleep[[ 'newid','Id',
 'ActivityDate',
 'TotalSteps',
 'TotalDistance',
 'TrackerDistance',
 'LoggedActivitiesDistance',
 'VeryActiveDistance',
 'ModeratelyActiveDistance',
 'LightActiveDistance',
 'SedentaryActiveDistance',
 'VeryActiveMinutes',
 'FairlyActiveMinutes',
 'LightlyActiveMinutes',
 'SedentaryMinutes',
 'Calories',
 'SleepDay',
 'TotalSleepRecords',
 'TotalMinutesAsleep',
 'TotalTimeInBed',
 'date_notime',
 'SleepStart',
 'SleepEnd']]
```

```python
#successfully matched activity entries to sleep entries
#however, sample is reduced to 413 as sleep df is smaller in size
activity_sleep
```

```python
#re-coding TotalSteps (continuous) into categories as one determinant of activity

#Sedentary is less than 5,000 steps per day; Low active is 5,000 to 7,499 steps per day;
#Somewhat active is 7,500 to 9,999 steps per day; Active is more than 10,000 steps per day;
#Highly active is more than 12,500 (Source:10000steps.org.au/)

bins = [0, 5000, 7499, 9999, 12500, float('inf')]
labels = ['Sedentary', 'Low active', 'Somewhat active', 'Active', 'Highly active']

activitySteps = pd.cut(activity_sleep['TotalSteps'], bins=bins, labels=labels, right=False)
activitySteps = pd.Series(activitySteps, name = 'activityinSteps')

```

```python
activity_sleep = pd.concat([activity_sleep, activitySteps ], axis = 1)
activity_sleep
```

```python
#total number of minutes per day: 24 x 60 = 1400 minutes
#creating variable: TotalActiveMinutes = VeryActiveMinutes + FairlyActive Minute + LightlyActiveMinutes

activity_sleep['TotalActiveMinutes'] = activity_sleep['VeryActiveMinutes'] + activity_sleep['FairlyActiveMinutes'] + activity_sleep['LightlyActiveMinutes']
activity_sleep
```

```python
#as there is no sleep quality data available, good sleep quality means spending most of the time asleep.
#lower sleep quality means the individual spent more time awake in bed.
#creating variable: SleepQuality = TotalMinutesAsleep / TotalTimeInBed x 100

activity_sleep['SleepQuality'] = (activity_sleep['TotalMinutesAsleep'] / activity_sleep['TotalTimeInBed']) * 100
activity_sleep['SleepQuality']
activity_sleep
```
